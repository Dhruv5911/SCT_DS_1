```python
# Task 04 — Traffic Accident Analysis
# Author: Chaitanya Kharb
# Internship Project: SmartInternz - Data Science
# -------------------------------------------------------------

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------------------------------------
# 1. Load the dataset
# -------------------------------------------------------------
df = pd.read_csv("data/traffic_accidents.csv")

# Display first 5 rows
df.head()
```

```python
# -------------------------------------------------------------
# 2. Basic Information about dataset
# -------------------------------------------------------------
print("Dataset Information:\n")
df.info()

print("\nSummary Statistics:\n")
print(df.describe(include='all'))
```

```python
# -------------------------------------------------------------
# 3. Checking for missing values
# -------------------------------------------------------------
print("Missing values per column:\n")
print(df.isnull().sum())

# If there are missing values, we can drop or fill them (not needed here)
# df = df.dropna()
```

```python
# -------------------------------------------------------------
# 4. EDA — Exploratory Data Analysis
# -------------------------------------------------------------

# 4.1 Count of accidents by severity
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Accident_Severity", palette="coolwarm")
plt.title("Accidents by Severity")
plt.show()
```

```python
# 4.2 Accidents by Weather Conditions
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Weather_Conditions", hue="Accident_Severity", palette="Set2")
plt.title("Accidents by Weather Condition and Severity")
plt.show()
```

```python
# 4.3 Accidents by Day of Week
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="Day_of_Week", order=[
    "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"
])
plt.title("Accidents by Day of Week")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4.4 Relationship between Time of Day and Accident Severity
# Convert Time to hour for better analysis
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))

plt.figure(figsize=(8,4))
sns.histplot(data=df, x="Hour", hue="Accident_Severity", multiple="stack", bins=12)
plt.title("Accident Distribution by Hour of Day")
plt.xlabel("Hour of Day")
plt.show()
```

```python
# 4.5 Accidents by Vehicle Type
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Vehicle_Type", hue="Accident_Severity", palette="muted")
plt.title("Accidents by Vehicle Type and Severity")
plt.show()
```

```python
# -------------------------------------------------------------
# 5. Casualties Analysis
# -------------------------------------------------------------
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Accident_Severity", y="Casualties", palette="coolwarm")
plt.title("Casualties by Accident Severity")
plt.show()
```

```python
# -------------------------------------------------------------
# 6. Summary Insights
# -------------------------------------------------------------
print("✅ Summary Insights:")
print("- Most accidents occurred under 'Clear' weather conditions, but rainy weather had more serious/fatal ones.")
print("- Majority of accidents happened during daylight hours.")
print("- 'Cars' are most frequently involved in accidents.")
print("- Casualties are slightly higher for 'Serious' and 'Fatal' accidents.")
```
