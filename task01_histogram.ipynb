```python
# ðŸ§  Task 01 â€” Visualization using World Bank Population Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Make sure folder exists for outputs
os.makedirs("outputs/figures", exist_ok=True)

# Load main dataset
file_path = "data/API_SP.POP.TOTL_DS2_en_csv_v2_130083.csv"
data = pd.read_csv(file_path, skiprows=4)   # World Bank datasets have 4 header rows
data.head()
```

```python
print(data.columns)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data['2020'], bins=30, kde=True)
plt.title("Global Population Distribution in 2020")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.savefig("outputs/figures/task01_population_hist.png")
plt.show()
```

```python
top10 = data[['Country Name', '2020']].sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='2020', y='Country Name', data=top10, palette='viridis')
plt.title("Top 10 Most Populated Countries (2020)")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig("outputs/figures/task01_top10_bar.png")
plt.show()
```
